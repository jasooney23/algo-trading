```python
# Python code for the plot

import numpy as np
import matplotlib.pyplot as plt

mu = 1
n = 50
dt = 0.1
x0 = 100
np.random.seed(1)

sigma = np.arange(0.8, 2, 0.2)

Wt = np.random.normal(0, np.sqrt(dt), size=(len(sigma), n))
x = np.exp(
    (mu - sigma ** 2 / 2) * dt
    + sigma * Wt.T
)

print(x[0])
x = np.vstack([np.ones(len(sigma)), x])
print(x[0])
x = x0 * x.cumprod(axis=0)
print(x[0])

plt.plot(x)
plt.legend(np.round(sigma, 2))
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title(
    "Realizations of Geometric Brownian Motion with different variances\n $\mu=1$"
)
plt.show()
```

```python
# Python code for the plot

import numpy as np
import matplotlib.pyplot as plt

mu = 1
n = 50
dt = 0.1
x0 = 100
sigma = 1
# np.random.seed(1)

x = np.ndarray((n,))
x[0] = x0

Wt = np.random.normal(0, np.sqrt(dt), size=(n))
# x = x0 * np.exp((mu - sigma ** 2 / 2) * dt + sigma * Wt)

for i in range(1, n):
    x[i] = x[i - 1] * np.exp((mu - sigma ** 2 / 2) * dt + sigma * Wt[i])
# x = x0 * np.exp((mu - sigma ** 2 / 2) * dt + sigma * Wt)

plt.plot(x)
# plt.legend(np.round(sigma, 2))
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title(
    "Realizations of Geometric Brownian Motion with different variances\n $\mu=1$"
)
plt.show()
```
